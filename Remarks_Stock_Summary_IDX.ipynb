```python
import pandas as pd
import sqlite3
```

# Menyiapkan data

```python
file = '../stock_summary_idx/2024/stock_summary_20240417.csv'
```

```python
cols = ['Stock Code', 'Remarks', 'Last Trading Date', 'Open Price', 'High', 'Low', 'Close']
db_sql = 'db_stock_summary.sqlite'
table_name = 'tabel'
```

```python
conn = sqlite3.connect(db_sql)
```

```python
pd.read_csv(file, usecols = cols ).to_sql(name=table_name,con=conn, if_exists='replace')
```

# 3.M Efek yang dapat ditransaksikan dan/atau dijaminkan dalam Transaksi Marjin


```python
q = f"""
    SELECT *
    FROM {table_name}
    WHERE SUBSTR(Remarks, 3,1) = "M";
    """
pd.read_sql_query(q, conn)
```

# 3.S Efek yang dapat ditransaksikan dan/atau dijaminkan dalam Transaksi Marjin dan Transaksi Short Selling.

```python
q = f"""
    SELECT *
    FROM {table_name}
    WHERE SUBSTR(Remarks, 3,1) = "S";
    """
pd.read_sql_query(q, conn)
```

# 4.- Saham yang tidak dapat diperdagangan pada sesi Pra-pembukaan (Pre-opening)


```python
q = f"""
    SELECT *
    FROM {table_name}
    WHERE SUBSTR(Remarks, 4,1) = "-";
    """
pd.read_sql_query(q, conn)
```

# 4.O Saham yang dapat diperdagangan pada sesi Pra-pembukaan (Pre-opening)

```python
q = f"""
    SELECT *
    FROM {table_name}
    WHERE SUBSTR(Remarks, 4,1) = "O";
    """
pd.read_sql_query(q, conn)
```

# Perusahaan bersih dari Notasi Khusus

```python
q = f"""
    SELECT *
    FROM {table_name}
    WHERE SUBSTR(Remarks, 19,12) = "------------";
    """
pd.read_sql_query(q, conn)
```

# 22.L Perusahaan Tercatat belum menyampaikan laporan keuangan

```python
q = f"""
    SELECT *
    FROM {table_name}
    WHERE SUBSTR(Remarks, 22,1) = "L";
    """
pd.read_sql_query(q, conn)
```

# 30.X Perusahaan Tercatat dicatatkan di Papan Pemantauan Khusus

```python
q = f"""
    SELECT *
    FROM {table_name}
    WHERE SUBSTR(Remarks, 30,1) = "X";
    """
pd.read_sql_query(q, conn)
```
