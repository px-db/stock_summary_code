```python
import pandas as pd
from Remarks import Remarks_idx as rmi
```

# Menyiapkan data

```python
file = '../stock_summary_idx/2024/stock_summary_20240417.csv'
cols = ['Stock Code', 'Remarks', 'Last Trading Date', 'Open Price', 'High', 'Low', 'Close']
```

```python
df = pd.read_csv(file, usecols = cols, index_col= 'Stock Code')
len(df)
```

# 3.M Efek yang dapat ditransaksikan dan/atau dijaminkan dalam Transaksi Marjin


```python
df[rmi().substr(df['Remarks'],3,1) == 'M']
```

# 3.S Efek yang dapat ditransaksikan dan/atau dijaminkan dalam Transaksi Marjin dan Transaksi Short Selling.

```python
df[rmi().substr(df['Remarks'],3,1) == 'S']
```

# 4.- Saham yang tidak dapat diperdagangan pada sesi Pra-pembukaan (Pre-opening)


```python
df[rmi().substr(df['Remarks'],4,1) == '-']
```

# 4.O Saham yang dapat diperdagangan pada sesi Pra-pembukaan (Pre-opening)

```python
df[rmi().substr(df['Remarks'],4,1) == 'O']
```
